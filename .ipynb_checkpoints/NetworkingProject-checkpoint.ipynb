```python
plot(iBali, bbox=(400,400), margin=20, 
    #vertex_size=20*deg,
    #vertex_size=10*np.log(deg),
     
    #vertex_size=100*cls,
    #vertex_size=100*np.log(cls),
     
    vertex_size=20*np.log(bet),
    #vertex_size=10*bet,
    ##vertex_size=np.sqrt(bet+1),
     
    vertex_color='red'
)
```

### 5.2.5 Edge Width

```python
ics = np.array(iBali.es['IC'])
```

```python
plot(iBali, bbox=(400,400), margin=20, 
    vertex_color='red',
    vertex_size=10,
    vertex_label=iBali.vs['vertexnames'],
    edge_color='blue',
    edge_curved=0.2,
    edge_width=1.5*ics #in pixels
)
```
